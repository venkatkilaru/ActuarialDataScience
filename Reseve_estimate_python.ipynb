```python
import pandas as pd
import numpy as np
import chainladder as cl
import os
```

```python
import matplotlib.pyplot as plt
```

```python
avg_claim_cost= pd.read_csv("ctp_avg_claim_cost.csv")
claim_cost= pd.read_csv("ctp_claim_cost.csv")
claim_count= pd.read_csv("ctp_claim_count.csv")

print(avg_claim_cost)
print(claim_cost)
print(claim_count)
```

```python
avg_claim_cost_triangle= cl.Triangle(avg_claim_cost, origin='Accident/underwriting year', development='Development year', columns=['sum(claim_cost)'], cumulative = True)

claim_cost_triangle= cl.Triangle(claim_cost, origin='Accident/underwriting year', development='Development year', columns=['sum(claim_cost)'],cumulative = True)

claim_count_triangle= cl.Triangle(claim_count, origin='Accident/underwriting year', development='Development year', columns=['sum(Value)'],cumulative = True)
```

```python
claim_cost_triangle
```

```python
claim_cost_triangle.link_ratio.heatmap()
```

```python

claim_cost_triangle.T.plot(
    marker='.', grid=True,
    title='plot of cumulative claims payment').set(
    xlabel='Development Period', ylabel='Cumulative Paid Loss');

plt.legend()
plt.figure(figsize=(20, 20))

```

```python
model = cl.Chainladder().fit(claim_cost_triangle)
claim_cost_full_triangle= model.full_triangle_
```

```python
claim_cost_full_triangle
```

```python
claim_cost_full_triangle.T.plot(
    marker='.', grid=True,
    title='plot of cumulative claims payment').set(
    xlabel='Development Period', ylabel='Cumulative Paid Loss');
```

```python
print(model.cdf_)
```

```python
print(model.ldf_)
```

```python
claim_cost_only_ibnr = claim_cost_full_triangle - claim_cost_triangle
claim_cost_only_ibnr
```
